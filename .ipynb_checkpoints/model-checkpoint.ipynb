### Import pandas to read dataset

```python
import pandas as pd

# read csv file and show first 5 rows
data = pd.read_csv('costumer_churn-dataset.csv')
data.head().T
```

### Drop some variable from dataset

```python
data.columns
```

```python
# Modify the dataset by removing the following attributes: 
# MonthlyCharges, OnlineSecurity, StreamingTV, InternetService and Partner.

remove_columns = ["MonthlyCharges", "OnlineSecurity", "StreamingTV", "InternetService", "Partner"]
new_data = data.drop(remove_columns, axis=1)
new_data.columns
```

### Check data set for missing values is exist

```python
# find missing value via info() function
new_data.info()
```

### Descriptive analysis with special function

```python
# present appropriate data visualisation and descriptive statistics
new_data.describe(include="all").T
```

### Descrip numeric and categorical variables to good visulitation and modelling

```python
new_data.TotalCharges = pd.to_numeric(new_data.TotalCharges, errors='coerce')

new_data = new_data.drop('customerID', axis = 1)
new_data.dropna(inplace = True)
```

```python
new_data['Churn'].replace(to_replace='Yes', value=1, inplace=True)
new_data['Churn'].replace(to_replace='No',  value=0, inplace=True)
```

```python
last_data = pd.get_dummies(new_data)
last_data.head()

# save last_data to csv
last_data.to_csv('modify_churn_dataset.csv')
```

```python
# Pearson Corr.
last_data.corr()
```

```python
# Importing and configuration of Spark Session
# Warning: Java 8 it must installed in OS
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
spark = SparkSession.builder.appName("Churn Prediction").getOrCreate()
```

```python
# read csv file with spark
df = spark.read.csv('modify_churn_dataset.csv', inferSchema=True, header=True)
df.printSchema()
```

```python
from pyspark.ml.feature import VectorAssembler
```

```python
df_feature = ['SeniorCitizen', 'tenure', 'TotalCharges',
 'gender_Female', 'gender_Male', 'Dependents_No', 'Dependents_Yes',
 'PhoneService_No', 'PhoneService_Yes', 'MultipleLines_No', 'MultipleLines_No phone service',
 'MultipleLines_Yes', 'OnlineBackup_No', 'OnlineBackup_No internet service', 'OnlineBackup_Yes',
 'DeviceProtection_No', 'DeviceProtection_No internet service', 'DeviceProtection_Yes',
 'TechSupport_No', 'TechSupport_No internet service', 'TechSupport_Yes', 'StreamingMovies_No',
 'StreamingMovies_No internet service', 'StreamingMovies_Yes', 'Contract_Month-to-month',
 'Contract_One year', 'Contract_Two year', 'PaperlessBilling_No', 'PaperlessBilling_Yes',
 'PaymentMethod_Bank transfer (automatic)','PaymentMethod_Credit card (automatic)',
 'PaymentMethod_Electronic check','PaymentMethod_Mailed check']
```

```python
assembler = VectorAssembler(inputCols=df_feature, outputCol='new_df_features')
```

```python
output = assembler.transform(df)
```

```python
model_df = output.select('new_df_features', 'Churn')
model_df.show()
```

```python
training_df, test_df = model_df.randomSplit([0.7,0.3])
```

```python
print(training_df.count())
print(test_df.count())
```

```python
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
df_classifier = DecisionTreeClassifier(labelCol="Churn", featuresCol="new_df_features").fit(training_df)
df_predictions = df_classifier.transform(test_df)
```

```python
df_predictions.show()
```

```python
df_accuracy = MulticlassClassificationEvaluator(labelCol='Churn',
                                                metricName='accuracy').evaluate(df_predictions)
df_accuracy
```

```python
df_precision = MulticlassClassificationEvaluator(labelCol='Churn',
                                                metricName='weightedPrecision').evaluate(df_predictions)

df_precision
```
